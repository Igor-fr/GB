# Урок 2. Анализ данных и проверка статистических гипотез.

### 1. В чём различие между зависимыми и независимыми выборками?

Выборки зависимы, если одному значению из выорки 1 можно поставить в соответствие только одно значение из выборки 2. В противном случае выборки независимы, если любому значению из выборки 1 в соответствие можно поставить любое значние из выборки 2.
Одним из численных характеристик зависимости выборок может являться коэффициент корреляции.

### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Параметрические критерии применяются когда известно распределение величины, т.е. его параметры - математическое ожидание и СКО генеральной совокупности, в противном случае применяется непараметрчиеские аналоги.
