# Урок 3. Построение модели классификации.

### 1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

micro - расчитывает метрики по реальному количеству правильных и неправильных предсказаний для всех классов вместе.
macro - расчитывает метрики для каждого класса независимо и берет их невзвешенное среднее.
weighted - расчитывает метрики для каждого класса независимо и берет их взвешенное среднее.
Соответственно данные варианты усреднения должны выбираться в зависимости от размера дисбаланса классов и важности верных предсказаний для малых классов. Если важно учесть точность предсказания для малых классов, то необходимо использовать macro, если же существеннее всего точность предсказания для крупных классов и малыми можно пренебречь - то weighted, если же классы пимерно равны между собой, то можно использовать micro.

### 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

xgboost не может сам работать с категориальными признаками, тогда как lightgbm и catboost необходимо просто указать названия таких признаков, но catboost значительно лучше работает с категориальными признаками.
lightgbm считается самым быстрым алгоритмом, за счет чего немного уступающим по точности.
